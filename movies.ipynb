```python
# Descobrindo os maiores orçamentos dos filmes do maior para o menor
# B -> Bilhões / M -> Milhões / K -> Mil 

def convert_to_str(budget):
    if isinstance(budget, (int, float)):
        
        if budget >= 1_000_000_000:  # Se for maior ou igual a 1 bilhão
            return f"{budget / 1_000_000_000:.1f}B".replace(".0", "")  # Converte para "M"
        
        elif budget >= 1_000_000:  # Se for maior ou igual a 1 milhão
            return f"{budget / 1_000_000:.1f}M".replace(".0", "")  # Converte para "M"
        
        elif budget >= 1_000:  # Se for maior ou igual a 1 milhar
            return f"{budget / 1_000:.1f}K".replace(".0", "")  # Converte para "K"
        
        else:
            return str(float(budget))  #se for menor que 1000 Mantém como float
    return str(budget)  # Se já for string, retorna sem alteração


budget_df['budget'] = budget_df['budget'].apply(convert_to_str) # Aplicando a conversão à coluna "budget"

display(budget_df[['Title', 'budget', 'Year']].head(20))
```

```python
# Função que retorna a string que mais se repete em uma coluna de um DataFrame.

def string_mais_repetida(df, coluna):
    
    # Verifica se a coluna existe no DataFrame
    if coluna not in df.columns:
        raise ValueError(f"A coluna '{coluna}' não existe no DataFrame.")

    # Conta a frequência de cada valor na coluna
    contagem = df[coluna].value_counts()
    print(contagem)

    # Retorna a string mais repetida e sua contagem
    string_mais_comum = contagem.idxmax()
    contagem_mais_comum = contagem.max()

    return string_mais_comum, contagem_mais_comum
```

```python
#Tratando os dados da coluna 'directors', type lista -> str.

filmes_df['directors'] = filmes_df['directors'].fillna('[]') # Substituir NaN por listas vazias
filmes_df['directors'] = filmes_df['directors'].apply(ast.literal_eval)# Converte strings que representam listas em listas reais
filmes_df['directors'] = filmes_df['directors'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None) # Extrai o primeiro nome da lista, se a lista não estiver vazia
```

```python
#Diretores com mais filmes

string, contagem = string_mais_repetida(filmes_df, 'directors')
print(f'\nO diretor com mais filmes desse dataset é {string}, com {contagem} filmes registrados no imdb')

```

```python
filmes_df['Languages'] = filmes_df['Languages'].fillna('[]') # Substituir NaN por listas vazias
filmes_df['Languages'] = filmes_df['Languages'].apply(ast.literal_eval)# Converte strings que representam listas em listas reais
filmes_df['Languages'] = filmes_df['Languages'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None) # Extrai o primeiro nome da lista, se a lista não estiver vazia
```

```python
# Frequência e Percenteual de quantas linguagens são mais presentes nos filmes

contagem = filmes_df['Languages'].value_counts()
percentual = filmes_df['Languages'].value_counts(normalize = True) * 100

percentual_df = pd.DataFrame({
    'Frequência': contagem,
    'Porcentagem (%)': percentual
})
percentual_df.head(20) #Porcentagem e qunatidade de filmes em cada língua ou combinação delas
```

```python
#Filmes na Língua Portuguesa mais premiados

titles = filmes_df.loc[filmes_df['Languages'] == 'Portuguese', 'Title']
years = filmes_df.loc[filmes_df['Languages'] == 'Portuguese', 'Year']
directors = filmes_df.loc[filmes_df['Languages'] == 'Portuguese', 'directors']
oscar = filmes_df.loc[filmes_df['Languages'] == 'Portuguese', 'oscars']

filmesPT_df = pd.DataFrame({
    'Títulos': titles,
    'Diretores': directors,
    'Ano': years,
    'Oscars': oscar
})
filmesPT_df.sort_values(by='Oscars', ascending=False).head(8)
```

```python
#Filmes em Lang-PT mais recentes
filmesPT_df.sort_values(by='Ano', ascending=False).head(15) 
```

```python
#Filmes brasileiros que ganharam ao menos 1 oscar ordenados por ano

# Garantir que a coluna 'Oscars' seja numérica
filmesPT_df['Oscars'] = pd.to_numeric(filmesPT_df['Oscars'], errors='coerce').fillna(0).astype(int)

oscarsBR_df = pd.DataFrame(columns=filmesPT_df.columns)

for i in filmesPT_df.index:
    if filmesPT_df.loc[i, 'Oscars'] > 0:  # Verifica se o filme ganhou Oscar
        oscarsBR_df = pd.concat([oscarsBR_df, filmesPT_df.loc[[i]]], ignore_index=True)
            
oscarsBR_df.sort_values(by='Ano', ascending=False)
```

```python
# Transformando os votos em float64 para fazer operações
def convert_votes(votes):
    if isinstance(votes, str):
        votes = votes.replace(",", "")  
        if "K" in votes:
            return float(votes.replace("K", "")) * 1_000  
        elif "M" in votes:
            return float(votes.replace("M", "")) * 1_000_000  
    return float(votes) if votes else 0  

filmes_df["Votes"] = filmes_df["Votes"].apply(convert_votes)

print(filmes_df["Votes"].dtype) 
```

```python
#Os 20 filmes mais bem avaliados do dataset (nota x quantidade de votos)

filmes_df['Score'] = filmes_df.Rating * filmes_df.Votes
filmes_df['Votes'] = filmes_df['Votes'].apply(convert_to_str)
display(filmes_df[['Title', 'Rating', 'Votes', 'Year', 'oscars', 'directors', 'Score']].sort_values(by='Score', ascending=False).head(20))
```

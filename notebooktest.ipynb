```python
def string_mais_repetida(df, coluna):
    """
    Retorna a string que mais se repete em uma coluna de um DataFrame.
    """
    # Verifica se a coluna existe no DataFrame
    if coluna not in df.columns:
        raise ValueError(f"A coluna '{coluna}' não existe no DataFrame.")

    # Conta a frequência de cada valor na coluna
    contagem = df[coluna].value_counts()
    print(contagem)

    # Retorna a string mais repetida e sua contagem
    string_mais_comum = contagem.idxmax()
    contagem_mais_comum = contagem.max()

    return string_mais_comum, contagem_mais_comum
```

```python
#Diretores com mais filmes

string, contagem = string_mais_repetida(filmes_df_sort, 'directors')
print(f'\nO diretor com mais filmes desse dataset é {string}, com {contagem} filmes registrados no imdb')

```

```python
venda = {'data': ['14/12/2015', '16/06/2021'],
         'valor': [500.90, 300.50],
         'produto': ['Micro-ondas', 'Liquidificador'],
         'qtde': [129, 547]
}

vendas_df = pd.DataFrame(venda)
```

```python
def string_mais_repetida(df, coluna):
    """
    Retorna a string que mais se repete em uma coluna de um DataFrame.

    Parâmetros:
    df (pd.DataFrame): O DataFrame que contém os dados.
    coluna (str): O nome da coluna onde a string será contabilizada.

    Retorna:
    str: A string que mais se repete.
    int: A contagem de vezes que a string aparece.
    """
    # Verifica se a coluna existe no DataFrame
    if coluna not in df.columns:
        raise ValueError(f"A coluna '{coluna}' não existe no DataFrame.")

    # Conta a frequência de cada valor na coluna
    contagem = df[coluna].value_counts()
    print(contagem)

    # Retorna a string mais repetida e sua contagem
    string_mais_comum = contagem.idxmax()
    contagem_mais_comum = contagem.max()

    return string_mais_comum, contagem_mais_comum
```

```python
string, contagem = string_mais_repetida(filmes_df, 'writers')

print(f"A string que mais se repete é '{string}' com {contagem} ocorrências.")
```
